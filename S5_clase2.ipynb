# HTML y XML: raspado web

Python tiene varias bibliotecas muy útiles para leer datos de páginas web escritas en HTML o XML. Las más comunes son:

- lxml
- Beautiful Soup
- html5lib

De estas, lxml es la más rápida, por lo que se usa mucho cuando los archivos están bien estructurados.
En cambio, si el archivo HTML o XML está mal escrito o tiene errores (algo común en páginas web reales), Beautiful Soup o html5lib pueden ser más adecuadas, ya que son más tolerantes y fáciles de usar.

Estas herramientas permiten extraer información estructurada de páginas web de forma eficiente y automatizada.

### Conversión a otros tipos comunes en `pandas`

Debemos limpiar la columna 'País (o territorio dependiente)'

```python
set(df['País (o territorio dependiente)'])
```

```python
paises_filtrados = []

for pais in df['País (o territorio dependiente)']:
    if "(" not in pais and ")" not in pais:
        paises_filtrados.append(pais)
    
```

```python
paises_filtrados
```

¿Qué hacemos para no perder el valor de su población?

La función **zip()** en Python sirve para emparejar elementos de múltiples listas (o iterables) en paralelo, creando tuplas con los elementos correspondientes de cada uno.

```python
paises_filtrados = []
poblaciones_filtradas = []

for pais, poblacion in zip(df['País (o territorio dependiente)'], 
                           df['Población']):
    if "(" not in pais and ")" not in pais:
        paises_filtrados.append(pais)
        poblaciones_filtradas.append(poblacion)
```

```python
paises_filtrados
```

```python
import re

paises_limpios = []

for pais in paises_filtrados:
    paises_limpios.append(re.sub(r"\[\d+\]|\u200b", "", pais).strip())
```

```python
paises_limpios
```

```python
df = pd.DataFrame({'Población':poblaciones_filtradas},
             index = paises_limpios)
df
```

```python
df1 = df.head(20)
```

```python
plt.figure(figsize=(7,7), dpi=250) 
plt.barh(df1.index, df1['Población'])
plt.title('20 Paises más poblados')
plt.show()
```

# XML

El lenguaje de marcado extensible (XML) permite definir y almacenar datos de forma compartible. XML admite el intercambio de información entre sistemas de computación, como sitios web, bases de datos y aplicaciones de terceros.

#### Etiquetas XML

Los símbolos de marcado, denominados etiquetas en XML, se utilizan para definir los datos. Por ejemplo, para representar los datos de una librería, puede crear etiquetas como < libro >, < título > y < autor >. El documento XML de un solo libro tendría el siguiente contenido:

< libro >

< título >Introducción a Amazon Web Services</título>

< autor >Mark Wilkins< /autor >

< /libro >

Las etiquetas ofrecen una sofisticada codificación de datos para integrar los flujos de información en diferentes sistemas.

```python
ruta = 'https://raw.githubusercontent.com/jamc88/AD-MCMAI/refs/heads/main/Datos/bestsellers_mexico.xml'
df = pd.read_xml(ruta) 
df.head()
```

La función **.countplot()** de Seaborn se usa para visualizar el conteo de observaciones en cada categoría de una variable categórica.

```python
plt.figure(figsize=(5,3), dpi=250)
sns.countplot(data=df, y='Genre', color='cadetblue') 
plt.show()
```

```python
orden = df['Genre'].value_counts().index
```

```python
plt.figure(figsize=(5,3), dpi=250)
sns.countplot(data=df, y='Genre', order=orden, color='cadetblue') 
plt.show()
```
