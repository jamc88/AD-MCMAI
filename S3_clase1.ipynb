## Población de México por año 

```python
ruta_poblacion='https://raw.githubusercontent.com/jamc88/AD-MCMAI/refs/heads/main/Datos/poblacion.csv?token=GHSAT0AAAAAADEUM34CZT2D6HDQSWWMEBBY2CIMQCA'
```

```python
poblacion = pd.read_csv(ruta_poblacion)
```

```python
poblacion.head()
```

```python
poblacion['1990'] = poblacion['H_1990'] + poblacion['M_1990']
poblacion['2000'] = poblacion['H_2000'] + poblacion['M_2000']
poblacion['2010'] = poblacion['H_2010'] + poblacion['M_2010']
poblacion['2020'] = poblacion['H_2020'] + poblacion['M_2020']
```

```python
poblacion = poblacion.drop(['H_1990','M_1990','H_2000',
                            'M_2000','H_2010','M_2010','H_2020',
                            'M_2020','Regiones'],axis=1)
```

```python
poblacion = poblacion.set_index("Estado")

poblacion.head()
```

```python
F=['1990','2000','2010','2020']
C=[poblacion['1990'].sum(),poblacion['2000'].sum(),
   poblacion['2010'].sum(),poblacion['2020'].sum()]
```

```python
pob = pd.DataFrame({'Poblacion' : C}, index=F)
```

```python
pob
```

```python
plt.bar(pob.index, pob['Poblacion'], color = 'blue', alpha=0.6)
plt.title('Poblacion por año ')
plt.ylabel('Poblacion (100,000,000)')
plt.xlabel('Año')
plt.show()
```

### Correlación

```python
contam = pd.read_excel('contam2022.xlsx')
```

```python
contam
```

```python
contam.columns = ['Fecha','ozono','dioxido_azufre',
                  'dioxido_nitrogeno','monoxido_carbono',
                  'PM10']
```

```python
contam
```

```python
contam = contam.set_index("Fecha")
```

```python
contam
```

El método **corr** del objeto Series calcula la correlación de los valores superpuestos, no nulos y alineados por índice en dos series.

```python
contam['ozono'].corr(contam['PM10'])
```

Los métodos **corr** y **cov** del objeto DataFrame, por otro lado, devuelven una correlación completa o una matriz de covarianza como un dataframe

```python
contam.corr()
```

De forma similar, **cov** calcula la covarianza:

```python
contam.cov()
```

Utilizando el método **corrwith** se pueden calcular correlaciones por pares entre las columnas de un Dataframe 

```python
contam.corrwith(contam['PM10'])
```
