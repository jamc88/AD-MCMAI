## Ozono (O3) por zona

```python
df_O3 = df[['Noroeste Ozono',
            'Noreste Ozono',
            'Centro Ozono',
            'Suroeste Ozono',
            'Sureste Ozono']]

df_O3.columns=['Noroeste','Noreste','Centro','Suroeste','Sureste']

df_O3
```

```python
df_O3.info()
```

```python
df_O3.isna().sum()
```

La mediana es una medida de tendencia central robusta, ideal para análisis de datos ambientales, financieros o clínicos, donde los outliers son comunes.

```python
for col in df_O3.columns:
    df_O3[col] = df_O3[col].fillna(df_O3[col].median())
    
```

```python
df_O3.isna().sum()
```

```python
df_O3.describe()
```

### Ozono por zona en una semana

Necesitamos agregar la hora a nuestro indice temporal.

La función **pd.to_timedelta()** de pandas convierte datos numéricos o cadenas en objetos de tipo Timedelta, que representan una diferencia de tiempo (por ejemplo, 3 horas, 15 minutos, 2 días, etc.).

*pd.to_timedelta(arg, unit='s')*

*arg*: número, lista, Series o string con duración (ej. '2 days', '00:30:00')

*unit*: (opcional) unidad de tiempo si arg es numérico

- 's': segundos
- 'm': minutos
- 'h': horas
- 'D': días

En muchos conjuntos de datos ambientales, las horas están registradas del 1 al 24, pero pandas cuenta horas desde cero (00:00) hasta 23:00.

```python
df['Hora']-1
```

```python
df_O3.index = df_O3.index + pd.to_timedelta(df['Hora'] - 1, unit='h')
```

```python
df_O3.index
```

```python
O3_periodo = df_O3.loc['2017-01-01':'2017-01-07']
```

```python
plt.figure(figsize=(14, 6),dpi=150)
for col in O3_periodo.columns:
    sns.lineplot(data=O3_periodo, x=O3_periodo.index, y=col, label=col)

plt.title('Concentraciones de $O_3$ por zona')
plt.xlabel('Fecha y hora')
plt.ylabel('Concentración de $O_3$ $(\mu g/m^3)$')
plt.legend(title='Estación')
plt.grid(True)
plt.tight_layout()
plt.tick_params(labelsize=14)
plt.show()
```

### ¿Cuál fue el promedio diario de las concentraciones de ozono por zona?

```python
O3_a = pd.DataFrame(df_O3.resample('D').mean())
O3_a
```

# Discretización

Los datos continuos se suelen discretizar o, lo que es lo mismo, separar en **categorías** para su análisis.

```python
ruta = 'https://raw.githubusercontent.com/jamc88/AD-MCMAI/refs/heads/main/Datos/Corredores.csv'

run = pd.read_csv(ruta)
```

```python
run
```

```python
run.describe()
```

Supongamos que queremos agrupar las edades de los corredores en grupos
- menores de 18 años
- 18 a 29 años
- 30 a 39 años
- 40 a 49 años
- 50 a 59 años
- mayores de 60 años

Para ello utilizamos **pandas.cut()**:

```python
edades = run['Edad']
```

```python
bins = [0,17,29,39,49,59,90]
```

```python
categorias = pd.cut(edades,bins)
```

```python
categorias
```

```python
pd.value_counts(categorias)
```

En la representación de cadena de texto de un intervalo, un paréntesis significa que el lado está abierto, mientras que el corchete significa que está cerrado. Se puede cambiar el lado que está cerrado pasando **right=False**:

```python
pd.cut(edades,bins,right=False)
```

Se puede anular el etiquetado de contenedor predeterminado basado en intervalos pasando una lista o un array a la opción labels:

```python
nombres_grupos = ['<18','18 a 29 años', 
                  '30 a 39 años', 
                  '40 a 49 años', 
                  '50 a 59 años', 
                  '≥60 años']
```

```python
pd.cut(edades,bins,labels = nombres_grupos)
```

```python
run['grupo_edad'] = pd.cut(edades, bins, labels = nombres_grupos)
```

```python
run.head()
```

```python
plt.figure(figsize=(10,5), dpi=250)
sns.barplot(data=run, x='Tiempo', y='grupo_edad', hue='Genero')
plt.tick_params(labelsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 6),dpi=150)
sns.boxplot(data=run, x='grupo_edad', y='Tiempo', hue='Genero')
plt.title('Gráfico de caja por grupo de edad y género')
plt.xlabel('Grupo de edad')
plt.ylabel('Tiempo (min)')
plt.grid(True)
plt.tight_layout()
plt.tick_params(labelsize=12)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, sharex=True, figsize=(10, 6),
                         height_ratios=[1, 3])

sns.boxplot(data=run, x='Edad', ax=axes[0], color='skyblue')
axes[0].set_title('Boxplot y distribución de edades')
axes[0].set(xlabel='')

sns.histplot(data=run, x='Edad', bins=30, kde=True, ax=axes[1], 
             color='lightgreen')
axes[1].set_xlabel('Edad')
axes[1].set_ylabel('Frecuencia')

plt.tick_params(labelsize=12)
plt.tight_layout()
plt.show()
```

| Parte                  | Significado                                                                                                                                                       |
| ---------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| `fig, axes =`          | Crea la figura (`fig`) y una lista o arreglo de ejes (`axes`) sobre los cuales se dibujan los gráficos.                                                           |
| `plt.subplots(...)`    | Función de Matplotlib para crear una figura con múltiples subplots.                                                                                               |
| `nrows=2`              | Número de filas de subplots (uno arriba y otro abajo).                                                                                                           |
| `sharex=True`          | Los subplots comparten el mismo eje X (útil para comparar variables).                                                                         |
| `height_ratios=[1, 3]` | Define las proporciones de altura de cada subplot: el de arriba tendrá 1 parte de alto y el de abajo 3 partes (es decir, el segundo subplot será más grande). |
