# Carga de datos y almacenamiento

#### Funciones de carga de datos de texto y binarios en pandas.

| **Función**        | **Descripción**                                                                                           |
|--------------------|---------------------------------------------------------------------------------------------------------------------|
| `read_csv()`       | Lee archivos delimitados por comas (CSV) desde disco, URL o buffer. Muy común para datos tabulares simples.        |
| `read_fwf()`       | Lee archivos de texto con columnas de ancho fijo. Útil cuando los datos no tienen delimitadores.                   |
| `read_clipboard()` | Lee datos copiados al portapapeles como si fueran un CSV. Ideal para copiar tablas desde Excel o páginas web.      |
| `read_excel()`     | Lee hojas de cálculo Excel (.xls, .xlsx). Soporta múltiples hojas. Usa `openpyxl` o `xlrd` como motores.            |
| `read_hdf()`       | Lee archivos HDF5 escritos con `pandas.HDFStore`. Eficiente para grandes volúmenes de datos jerárquicos.            |
| `read_html()`      | Extrae todas las tablas de una página HTML (como tablas de Wikipedia) y devuelve una lista de DataFrames.          |
| `read_json()`      | Lee archivos JSON o cadenas JSON válidas. Admite JSON anidados con `orient='records'` o `orient='split'`.           |
| `read_feather()`   | Lee archivos en formato Feather, optimizado para velocidad y compatibilidad con Apache Arrow.                       |
| `read_orc()`       | Lee archivos binarios en formato ORC (Optimized Row Columnar), usado en sistemas distribuidos como Hive.            |
| `read_parquet()`   | Lee archivos Parquet, un formato columnar eficiente muy usado en Big Data y Spark.                                 |
| `read_pickle()`    | Carga objetos serializados por `pandas.to_pickle()` con `pickle`. Útil para guardar DataFrames entre sesiones.     |
| `read_sas()`       | Lee archivos `.sas7bdat` o `.xpt` del software estadístico SAS.                                                     |
| `read_spss()`      | Carga archivos `.sav` de SPSS como DataFrames. Requiere `pyreadstat`.                                               |
| `read_sql()`       | Ejecuta consultas SQL sobre una conexión de base de datos (usando SQLAlchemy) y devuelve los resultados como DataFrame. |
| `read_sql_table()` | Carga una tabla completa desde una base de datos relacional vía SQLAlchemy, sin necesidad de escribir SQL manual.   |
| `read_stata()`     | Lee archivos `.dta` del software Stata. Puede importar etiquetas de variables y valores.                           |

En estos casos, se puede pasar una expresión regular como delimitador para pandas.read_csv. Esto puede expresarse con la expresión regular **\s+**

| **Símbolo** | **Significado**                                                                 |
|-------------|----------------------------------------------------------------------------------|
| `\s`        | Representa **un carácter de espacio en blanco** (espacio, tabulación, salto de línea, etc.). |
| `+`         | Significa "**una o más veces**" del carácter anterior.                          |

Entonces, `\s+` significa:  
“Uno o más espacios en blanco consecutivos”


```python
ruta = "https://raw.githubusercontent.com/jamc88/AD-MCMAI/refs/heads/main/Datos/signals.txt"
df = pd.read_csv(ruta, sep='\s+')
df.head()
```

### Otros patrones comunes en expresiones regulares

| **Patrón** | **Descripción**                                         |
|------------|----------------------------------------------------------|
| `\d`       | Cualquier dígito (`0–9`)                                 |
| `\D`       | Cualquier carácter que **no** sea un dígito              |
| `\w`       | Cualquier carácter alfanumérico (`a–z`, `A–Z`, `0–9`, `_`) |
| `\W`       | Cualquier carácter que **no** sea alfanumérico           |
| `\s`       | Espacio en blanco (espacio, tab, salto de línea, etc.)   |
| `\S`       | Cualquier carácter que **no** sea espacio en blanco      |
| `.`        | Cualquier carácter excepto salto de línea (`\n`)         |
| `^`        | Inicio de línea                                          |
| `$`        | Fin de línea                                             |
| `*`        | Cero o más repeticiones del patrón anterior              |
| `+`        | Una o más repeticiones del patrón anterior               |
| `?`        | Cero o una repetición del patrón anterior                |
| `{n}`      | Exactamente `n` repeticiones                             |
| `{n,}`     | `n` o más repeticiones                                   |
| `{n,m}`    | Entre `n` y `m` repeticiones                             |
| `[abc]`    | Cualquier carácter entre `a`, `b`, o `c`                 |
| `[^abc]`   | Cualquier carácter **excepto** `a`, `b` o `c`            |
| `(a\|b)`    | Coincide con `a` **o** con `b`                           |

 ### Algunos argumentos de la función pandas.read_csv.

| **Argumento**      | **Descripción**                                                                                   |
|--------------------|------------------------------------------------------------------------------------------------------------|
| `path`             | Ruta del archivo (local o URL). También puede ser un objeto tipo archivo o buffer.                        |
| `sep` / `delimiter`| Separador de campos. Por defecto `','`. Usa `\t` para TSV, o una expresión regular como `'\s+'` para espacios. |
| `header`           | Índice de fila a usar como nombres de columnas. Usa `header=None` si no hay encabezado.                   |
| `index_col`        | Columna(s) a usar como índice del DataFrame. Puede ser nombre, número o lista.                            |
| `names`            | Lista de nombres de columnas. Útil si no hay encabezado (`header=None`).                                   |
| `skiprows`         | Número de filas (o lista de índices) a omitir al inicio. Ideal para saltar metadatos o comentarios.        |
| `na_values`        | Lista de strings que serán interpretados como NA/NaN. Ej: `["n/a", "NA", "--"]`                            |
| `keep_default_na`  | Si `False`, desactiva los valores NA predeterminados (como `'NA'`, `'null'`, etc.).                       |
| `comment`          | Carácter que indica que el resto de la línea debe ignorarse como comentario. Ej: `comment='#'`            |
| `parse_dates`      | Intenta convertir columnas a datetime. Usa `True`, una lista de columnas o combinaciones de columnas.     |
| `keep_date_col`    | Si `True`, mantiene las columnas originales usadas para crear fechas.                                      |
| `converters`       | Diccionario de funciones para aplicar a columnas específicas. Ej: `{"col1": lambda x: x.strip()}`         |
| `dayfirst`         | Si `True`, interpreta fechas ambiguas como día/mes/año. (por ej. 10/11/2022 → 11 de octubre)              |
| `date_parser`      | Función personalizada para convertir fechas. Se usa con `parse_dates`.                                     |
| `nrows`            | Número de filas a leer desde el inicio (sin contar encabezado).                                            |
| `iterator`         | Devuelve un iterador (`TextFileReader`) para leer por fragmentos (útil con `chunksize`).                   |
| `chunksize`        | Tamaño de los bloques (en filas) que se leen por iteración. Ideal para archivos muy grandes.              |
| `skip_footer`      | Número de filas a ignorar desde el final del archivo. Requiere `engine='python'`.                         |
| `verbose`          | Muestra información adicional durante la carga del archivo (útil para depuración).                         |
| `encoding`         | Codificación del archivo. Por ejemplo: `"utf-8"`, `"latin1"`, `"cp1252"`                                   |
| `squeeze`          | Si el archivo tiene una sola columna, la devuelve como `Series` en vez de `DataFrame`.                    |
| `thousands`        | Separador de miles (como `','` o `'.'`) para convertir strings a números correctamente.                    |
| `decimal`          | Separador decimal, por defecto `"."`. Usa `","` para archivos europeos.                                   |
| `engine`           | Motor de lectura: `"c"` (rápido por defecto), `"python"` (más flexible), `"pyarrow"` (experimental).       |

Utilizamos el parámetro **`chunksize`** cuando trabajamos con conjuntos de datos grandes para dividir el archivo en fragmentos manejables. Este parámetro permite especificar el número de filas por bloque que se leerá en cada iteración.

Al procesar los datos por partes en lugar de cargar todo el archivo a la vez, se logra:

- Reducir el uso de memoria.
- Mejorar la eficiencia del código, especialmente en sistemas con recursos limitados.
- Facilitar el procesamiento en streaming o en paralelo de grandes volúmenes de datos.

Esto es particularmente útil al leer archivos con millones de registros o al aplicar transformaciones parciales antes de construir el DataFrame completo.


```python
ruta = "https://raw.githubusercontent.com/jamc88/AD-MCMAI/refs/heads/main/Datos/carpetasFGJ.csv"
df = pd.read_csv(ruta, chunksize=1_000)
```

```python
df
```

Cuando usamos el parámetro **`chunksize`** en pandas.read_csv(), la función no devuelve directamente un `DataFrame`, sino un objeto `TextFileReader`. Este objeto actúa como un *iterador* que permite recorrer el archivo por bloques del tamaño especificado.

Esto resulta especialmente útil cuando se desea procesar el archivo por partes, por ejemplo, para:

- Calcular estadísticas parciales (como conteos por categoría),
- Filtrar datos en tiempo real,
- O construir gradualmente un `DataFrame` sin cargar todo en memoria.

Además, podemos aplicar operaciones como `value_counts()` a una columna específica en cada fragmento, acumulando los resultados al iterar:


```python
ruta = "https://raw.githubusercontent.com/jamc88/AD-MCMAI/refs/heads/main/Datos/carpetasFGJ.csv"
df = pd.read_csv(ruta, chunksize = 1_000)

tot = pd.Series([], dtype='int64')
    
for piece in df:
    tot = tot.add(piece['delito'].value_counts(), fill_value=0)
    tot = tot.sort_values(ascending=False)

delitos = pd.DataFrame(tot,columns=['Frecuencia'])
delitos
```

```python
delitos = delitos.head(15).sort_values(by=['Frecuencia'])
```

```python
plt.figure(figsize=(7,9), dpi=150)
plt.barh(delitos.index, delitos['Frecuencia']
         ,facecolor='#6495ED', edgecolor='white')
plt.title('Delitos CDMX', fontsize = 15)
plt.grid(True)
plt.show()
```

## Datos JSON

**JSON**, que significa JavaScript Object Notation (notación de objetos de JavaScript), se ha convertido en uno de los formatos estándares para enviar datos mediante petición HTTP entre navegadores web y otras aplicaciones. 

JSON es código de Python perfectamente válido, con la excepción de su valor nulo null y otras menudencias (como no permitir comas al final de las listas). Los tipos básicos son objetos (dicccionarios), arrays (listas), cadenas de texto, números, valores booleanos y nulos. Todas las claves de un objeto deben ser cadenas de texto.

```python
import json
import requests

ruta = 'https://raw.githubusercontent.com/jamc88/AD-MCMAI/refs/heads/main/Datos/localidades_100.json'
response = requests.get(ruta)
response.raise_for_status()


localidades = json.loads(response.text)
```
